# Application of divergence of TV filter on an image 

This is a way to generate pretty cool images by application of simple gradient and divergence to images, based on the fact that most of the images contain local noise. This would not work on very smooth images.  
Total variation filter is the gradient at each point and its summation. A divergence of which gives us the resulting image. 

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image_path = "Samyak_photo_new.jpeg" # Insert an image of your choice
image = cv2.imread(image_path) 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert the image into black and white, can be done for rgb by appyling this individually to the three color images, but on black white this appears better 

plt.imshow(image, cmap='gray')
```

```python
grad_x = np.gradient(image, axis=1)
grad_y = np.gradient(image, axis=0)

epsilon = 0.001

grad_x=grad_x/(np.sqrt(grad_x**2+grad_y**2+epsilon**2))
grad_y=grad_y/(np.sqrt(grad_x**2+grad_y**2+epsilon**2))

image_f = -1*(np.gradient(grad_x, axis=1) + np.gradient(grad_y, axis=0))

plt.imshow(image_f, cmap='gray')
```

```python
epsilon = 1000

grad_x=grad_x/(np.sqrt(grad_x**2+grad_y**2+epsilon**2))
grad_y=grad_y/(np.sqrt(grad_x**2+grad_y**2+epsilon**2))

image_f = -1*(np.gradient(grad_x, axis=1) + np.gradient(grad_y, axis=0))

plt.imshow(image_f, cmap='gray')
```
